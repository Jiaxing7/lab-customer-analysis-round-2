5.Check and deal with NaN values.

```python
#check NaN values
file.isna().sum()
```

```python
#drop rows that have null values (only if there are very few)
# select 'is NaN' == 'False', means drop the 'Nan' == 'True', the rest are not NaN.
file = file[file['state'].isna()==False]
file.isna().sum()
```

```python
# to see months_since_last_claim value
file['months_since_last_claim'].value_counts()
```

```python
#it has a lot different numbers,and all close
# strategy: impute a value for the missing value (fill in a value that we choose or calculate)
#mean
mean_mslc = file['months_since_last_claim'].mean()
file['months_since_last_claim'] = file['months_since_last_claim'].fillna(mean_mslc)
file.isna().sum() 
```

```python
# to see the number_of_open_complaints value
file['number_of_open_complaints'].value_counts()
```

```python
#the most are 0.0, so assigned the missing value with 0
# use most common value to fillna
file['number_of_open_complaints'] = file['number_of_open_complaints'].fillna(0.0)
file.isna().sum()
```

```python
#fill unknown (object)
file['vehicle_class'] = file['vehicle_class'].fillna('unknown_class')
file['vehicle_size'] = file['vehicle_size'].fillna('unknown_size')
file.isna().sum()
```

```python
# to see the vehicle_type value
file['vehicle_type'].value_counts()
```

```python
# all are "A", so use "M" manual instead of NaN
file['vehicle_type'] = file['vehicle_type'].fillna('M')
file.isna().sum()
```

3.One Hot/Label Encoding (categorical).

```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
```

```python
X_cat = file.select_dtypes(include = object)
X_cat
```

```python
# drop customer
X_cat = X_cat.drop(['customer'],axis=1)
X_cat
```

```python
# just show "state"column
state_cat = X_cat[['state']]
state_cat
```

```python
from sklearn.preprocessing import OneHotEncoder

#drop the first value, show the rest
state_encoder = OneHotEncoder(drop='first').fit(state_cat)
print(state_encoder.categories_)     
```

```python
# same,use onehot Encoding chage categorical to number
#coverage
coverage_cat = X_cat[['coverage']]
coverage_encoder = OneHotEncoder(drop='first').fit(coverage_cat)
print(coverage_encoder.categories_) 
```

```python
coverage_encoded = coverage_encoder.transform(coverage_cat).toarray()
print(coverage_encoded)    
coverage_onehot_encoded = pd.DataFrame(coverage_encoded,columns=[ 'Extended', 'Premium'])   
coverage_onehot_encoded
```

```python
#education
education_cat = X_cat[['education']]
education_encoder = OneHotEncoder(drop='first').fit(education_cat)
print(education_encoder.categories_) 
```

```python
education_encoded = education_encoder.transform(education_cat).toarray()
print(education_encoded)    
education_onehot_encoded = pd.DataFrame(education_encoded,columns=[ 'College', 'Doctor', 'High School or Below', 'Master'])   
education_onehot_encoded
```

## Linear Regression

1.Train-test split.

```python
y = file['total_claim_amount']
y
```

```python
# train test split is the way ML generates its claim to fame: 
# we build the model on a portion of the data but we then validate it in 
# another "fresh" portion

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)    #20%from the whole train
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

2.Apply linear regression.

```python
#train/fit model
lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score

predictions = lm.predict(X_train)
r2_score(y_train, predictions)
```

```python
# But now we evaluate it in the TEST portion of the data, that we did not use for training.
# This way we know our model is genuinely guessing our donations, not just repeating the values it has seen in the training data


predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

```python
y_test[:5]
```

```python
predictions_test[:5]
```

### Model Validation

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions_test)
mse
#the mean of the squared differences between the predicted and actual values. 
#MSE = (1/n) * Σ(yi - ŷi)^2
#A lower MSE indicates a better fit between the model and the data.
```

```python
rmse = np.sqrt(mean_squared_error(y_test,predictions_test))
rmse
```

```python
y_test.mean()
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
mae = mean_absolute_error(y_test, predictions_test)
print(mae)
#the average absolute difference between the predicted values and the actual values of the dependent variable.
#MAE = (1/n) * Σ|yi - ŷi|
#A lower MAE indicates a better fit between the model and the data.
```

```python
rmse = math.sqrt(mse)
print(rmse)
#The RMSE is calculated by taking the square root of the mean of the squared differences between the predicted and actual values.
#RMSE = sqrt((1/n) * Σ(yi - ŷi)^2)
#A lower RMSE indicates a better fit between the model and the data.
```

```python
r2 = r2_score(y_test, predictions_test)
r2
#R2 ranges from 0 to 1
#value of 1 indicating that the regression model perfectly predicts the dependent variable based on the independent variables, 
#a value of 0 indicating that the model provides no predictive power.
```

Dealing with outliers.

```python
X
```

```python
sns.displot(x=X['customer_lifetime_value'])
plt.show()
```

```python
# removing the outliers from customer_lifetime_value - we don't always do it, only for good reasons
# but if we want to do it, this is how:
#X with Y together

iqr = np.percentile(X['customer_lifetime_value'],75) - np.percentile(X['customer_lifetime_value'],25)
upper_limit = np.percentile(X['customer_lifetime_value'],75) + 1.5*iqr
lower_limit = np.percentile(X['customer_lifetime_value'],25) - 1.5*iqr

X = X[(X['customer_lifetime_value']>lower_limit) & (X['customer_lifetime_value']<upper_limit)].copy()    
sns.displot(X['customer_lifetime_value'])
plt.show()
```
